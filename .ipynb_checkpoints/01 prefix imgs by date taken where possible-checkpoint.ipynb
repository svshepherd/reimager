# setup prefixes

```python
import os
from PIL import Image, ExifTags


#PATHNAME = r'F:\extended_desktop\ims'
PATHNAME = r'F:\extended_desktop\ims\_incoming'
#PATHNAME = r'F:\extended_desktop\ims\_incoming\10020926'
#PATHNAME = r'F:\extended_desktop\ims\_incoming\10120927'
#PATHNAME = r'F:\extended_desktop\ims\_incoming\10220928'
#PATHNAME = r'F:\extended_desktop\ims\_incoming\10320929'
#PATHNAME = r'F:\extended_desktop\ims\_incoming\10420930'



files = [f for f in os.listdir(PATHNAME) if os.path.isfile(os.path.join(PATHNAME, f))] 
files
```

```python
#[(k,v) for k,v in ExifTags.TAGS.items() if 'time' in v.lower()]

### example results:
# [(306, 'DateTime'),
#  (33434, 'ExposureTime'),
#  (34858, 'TimeZoneOffset'),
#  (34859, 'SelfTimerMode'),
#  (36867, 'DateTimeOriginal'),
#  (36868, 'DateTimeDigitized'),
#  (36880, 'OffsetTime'),
#  (36881, 'OffsetTimeOriginal'),
#  (36882, 'OffsetTimeDigitized'),
#  (37520, 'SubsecTime'),
#  (37521, 'SubsecTimeOriginal'),
#  (37522, 'SubsecTimeDigitized'),
#  (42082, 'CompositeImageExposureTimes'),
#  (50971, 'PreviewDateTime')]

### evidently I may also need to think about  (36881, 'OffsetTimeOriginal') and (36882, 'OffsetTimeDigitized'),
```

```python
for file in files:
    if file.endswith('.ARW'):
        continue
    if file[4]+file[7]=='--':
        newname = file[:4]+file[5:7]+file[8:]
        os.rename(os.path.join(PATHNAME, file), os.path.join(PATHNAME, newname))
        print(file, "---->", newname, "SUCCESSFUL")
    elif file.startswith('PXL_'):
        newname = file[4:]
        os.rename(os.path.join(PATHNAME, file), os.path.join(PATHNAME, newname))
        print(file, "---->", newname, "SUCCESSFUL")
    elif file.startswith('D')|file.startswith('P')|file.startswith('W'):
        try: 
            img = Image.open(os.path.join(PATHNAME, file))
            img_exif = img.getexif()
            ## [(k,v) for k,v in ExifTags.TAGS.items() if 'date' in v.lower()] ### 36867, 36868, 306
            ### evidently I also need to think about OffsetTime from GMT
            if 36867 in img_exif.keys():
                imname_pref = ''.join(img_exif[36867][:10].split(':'))+' '
            elif 36868 in img_exif.keys():
                imname_pref = ''.join(img_exif[36868][:10].split(':'))+' '
            else:
                imname_pref = ''.join(img_exif[306][:10].split(':'))+' '
            img.close()
            if imname_pref == file[:9]:
                print(file, "----=", file)
                try:
                    fileraw = file[9:].replace('.JPG','.ARW')
                    os.rename(os.path.join(PATHNAME, fileraw), os.path.join(PATHNAME, imname_pref + fileraw))
                    print(fileraw, "---->", imname_pref + fileraw, "SUCCESSFUL")
                except Exception as e:
                    print(e)
            else:
                print(file, "---->", imname_pref + file)
                os.rename(os.path.join(PATHNAME, file), os.path.join(PATHNAME, imname_pref + file))
                print(file, "---->", imname_pref + file, "SUCCESSFUL")
                try:
                    fileraw = file.replace('.JPG','.ARW')
                    os.rename(os.path.join(PATHNAME, fileraw), os.path.join(PATHNAME, imname_pref + fileraw))
                    print(fileraw, "---->", imname_pref + fileraw, "SUCCESSFUL")
                except Exception as e:
                    print(e + '.')
        except Exception as e:
            print(file, "----X", e)
    else: 
        print('  already processed or unanticipated name format: ', file)
```

# purge awr files where no jpg is retained

```python
import os
from PIL import Image, ExifTags

## goal here is to remove any .arw raw files where i didn't think the .jpg was worth saving

## confirm:
PATHNAME = r'F:\extended_desktop\ims\_incoming\10020926'

files = [f for f in os.listdir(PATHNAME) if os.path.isfile(os.path.join(PATHNAME, f))] 

raw_files = [f for f in files if f.lower().endswith('.arw')]
jpg_files = [f for f in files if f.lower().endswith('.jpg')]
jpg_stem = [f[:-4] for f in jpg_files]
raw_to_lose = [f for f in raw_files if f[:-4] not in jpg_stem]

display(f"Focused folder is {PATHNAME}")
len(raw_to_lose), len(raw_files)
```

```python
for f in raw_to_lose:
    print(f"removing {f}")
    os.remove( os.path.join(PATHNAME, f) )
```
